```python
try:
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    result = num1 / num2
except (ZeroDivisionError, ValueError):
    print("Error: Division by zero or invalid string input occurred.")
else:
    print(f"The result is: {result}")
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name          # Public property
        self.__age = age          # Private property (variable)

    def update_age(self, new_age):
        self.__age = new_age

    def display_details(self):
        print(f"Name: {self.name}, Age: {self.__age}")

# إنشاء كائن (Object)
p1 = Person("ASEEL", 20)
# تحديث العمر
p1.update_age(20)
# عرض البيانات
p1.display_details()
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

class Manager(Employee):
    def add_bonus(self, bonus):
        self.salary += bonus

    def display_info(self):
        print(f"Manager Name: {self.name}, Total Salary: {self.salary}")

# إنشاء كائن من Manager
mgr = Manager("Yo", 5000)
mgr.add_bonus(1000)
mgr.display_info()
```
