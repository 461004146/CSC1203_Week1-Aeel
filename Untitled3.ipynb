```python
sum_even = 0

for i in range(10):
    num = int(input("Enter a number: "))

    if num == 0:
        break

    if num < 0:
        continue

    if num % 2 == 0:
        sum_even += num   # augmented assignment

print("Final sum:", sum_even)
```

```python
num = int(input("Enter a number: "))

for i in range(1, 13):
    product = num * i

    if product % 3 == 0:
        continue

    if product > 50:
        break

    print(f"{num} x {i} = {product}")
```

```python
import random

secret = random.randint(1, 20)
attempts = 0

while True:
    guess = int(input("Guess a number between 1 and 20: "))

    if guess < 1 or guess > 20:
        print("Invalid guess! Try again.")
        continue

    attempts += 1   # augmented assignment

    if guess == secret:
        print("Correct! You guessed it.")
        print("Valid attempts:", attempts)
        break
    else:
        print("Wrong guess. Try again.")
```

```python
n = int(input("Enter a number: "))
count = 0

for i in range(1, n + 1):

    if i > 50:
        break

    if i % 8 == 0:
        continue

    if i % 4 == 0:
        count += 1

print("Count:", count)
```
