```python
import numpy as np

# 1. [span_0](start_span)إنشاء المصفوفة والتعرف على خصائصها[span_0](end_span)
A = np.arange(1, 21).reshape(4, 5)
print("Array A:\n", A)
print("Dtype:", A.dtype, "Dim:", A.ndim, "Shape:", A.shape, "Size:", A.size, "Itemsize:", A.itemsize)

# [span_1](start_span)تغيير شكل المصفوفة إلى (5, 4)[span_1](end_span)
B = A.reshape(5, 4)
print("\nArray B (Reshaped):\n", B)

# 2. [span_2](start_span)العمليات الحسابية والـ Broadcasting[span_2](end_span)
a = np.array([[0., 0., 0.],
              [10., 10., 10.],
              [20., 20., 20.],
              [30., 30., 30.]])
b = np.array([0., 1., 2.])

c = a + b
print("\nArray c (a + b):\n", c)

# [span_3](start_span)القناع المنطقي والعد[span_3](end_span)
mask = (c >= 22)
print("Boolean Mask (c >= 22):\n", mask)
print("Count values >= 22:", np.sum(mask))

# 3. [span_4](start_span)تحليل مصفوفة الدرجات[span_4](end_span)
grades = np.array([[87, 96, 70],
                  [100, 87, 90],
                  [94, 77, 90],
                  [100, 81, 82]])

print("\nOverall Stats:")
print(f"Sum: {grades.sum()}, Min: {grades.min()}, Max: {grades.max()}")
print(f"Mean: {grades.mean():.2f}, Std: {grades.std():.2f}, Var: {grades.var():.2f}")

print("Mean per column (axis=0):", grades.mean(axis=0))
print("Mean per row (axis=1):", grades.mean(axis=1))
print("Max per row (axis=1):", grades.max(axis=1))

# 4. [span_5](start_span)مصفوفة 2x3 وعمليات الجمع[span_5](end_span)
# ملاحظة: القيم المعطاة في السؤال تشكل مصفوفة 3x3، سأستخدمها كما وردت
arr4 = np.array([[2, 4, 6], 
                [1, 3, 5], 
                [7, 9, 11]])
print("\nMatrix 4 Stats:")
print("Sum of each row:", arr4.sum(axis=1))
print("Sum of each column:", arr4.sum(axis=0))
print("Overall mean:", arr4.mean())

# 5. [span_6](start_span)العمليات الحسابية البسيطة[span_6](end_span)
A5 = np.array([[10, 20, 30], 
               [40, 50, 60]])
B5 = np.array([1, 2, 3])

print("\nOperations on A5:")
print("Add B to A:\n", A5 + B5)
print("Multiply A by 2:\n", A5 * 2)
print("Subtract 5 from A:\n", A5 - 5)

# 6. [span_7](start_span)إيجاد أماكن القيم (Indices)[span_7](end_span)
grades6 = np.array([[70, 85, 90],
                   [88, 76, 95],
                   [60, 80, 72]])

print("\nIndex findings:")
print("Index of largest value (overall):", np.argmax(grades6))
print("Index of highest grade in each row:", np.argmax(grades6, axis=1))
print("Index of lowest grade in each column:", np.argmin(grades6, axis=0))
```
