```python
import math

# Function to calculate area
def calculate_area(radius):
    return math.pi * radius ** 2

# Lambda to calculate circumference
calculate_circumference = lambda r: 2 * math.pi * r

radius = float(input("Enter the radius of the circle: "))

area = calculate_area(radius)
circumference = calculate_circumference(radius)

print("Area:", round(area, 2))
print("Circumference:", round(circumference, 2))
```

```python
import re

# Function to sum numbers
def sum_numbers(numbers):
    return sum(numbers)

# Lambda to double the sum
double = lambda x: x * 2

text = input("Enter a string: ")

numbers = re.findall(r'\d+', text)
numbers = [int(num) for num in numbers]

total = sum_numbers(numbers)
doubled = double(total)

print("Numbers found:", numbers)
print("Sum:", total)
print("Doubled value:", doubled)
```

```python
import math

nums = input("Enter numbers separated by commas: ")
numbers = [float(x) for x in nums.split(",")]

# Function using lambda to square numbers
square = lambda x: x ** 2

squares = [square(n) for n in numbers]

max_square = max(squares)
sqrt_sum_squares = math.sqrt(sum(squares))

print("Squares:", squares)
print("Maximum square:", max_square)
print("Square root of sum of squares:", sqrt_sum_squares)
```

```python
import re
import math

password = input("Enter password: ")

# Check password strength
if (re.search(r'[A-Z]', password) and
    re.search(r'[a-z]', password) and
    re.search(r'[0-9]', password)):
    strength = "Strong"
else:
    strength = "Weak"

scores = list(map(float, input("Enter 5 scores separated by spaces: ").split()))

# Function to calculate average
def average(nums):
    return sum(nums) / len(nums)

avg = average(scores)

# Lambda (optional but required)
avg_lambda = lambda x: sum(x) / len(x)

rounded_avg = math.ceil(avg)

print("Password is:", strength)
print("Rounded Average:", rounded_avg)
```

```python
import re

# Function to count vowels
def count_vowels(text):
    vowels = "aeiouAEIOU"
    return sum(1 for char in text if char in vowels)

paragraph = input("Enter a paragraph: ")

# Find words starting with vowel
words = re.findall(r'\b[aeiouAEIOU]\w*', paragraph)

total_vowels = count_vowels(paragraph)

print("Words starting with vowel:", words)
print("Total vowels in paragraph:", total_vowels)
```
