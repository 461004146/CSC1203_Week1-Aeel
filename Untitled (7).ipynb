```python
import os

# 1 & 2. Create lab_data and subfolders
base_dir = "lab_data"
subfolders = ["input", "output", "logs"]

if not os.path.exists(base_dir):
    os.makedirs(base_dir)
    print(f"Created directory: {base_dir}")

for sub in subfolders:
    path = os.path.join(base_dir, sub)
    if not os.path.exists(path):
        os.makedirs(path)
        print(f"Created subfolder: {path}")

# 3. List all items and identify if they are files or directories
print("\nContents of lab_data:")
for item in os.listdir(base_dir):
    full_path = os.path.join(base_dir, item)
    if os.path.isdir(full_path):
        print(f"{item} is a Directory")
    else:
        print(f"{item} is a File")

# 4. Rename 'logs' to 'history'
old_name = os.path.join(base_dir, "logs")
new_name = os.path.join(base_dir, "history")
if os.path.exists(old_name):
    os.rename(old_name, new_name)
    print(f"\nRenamed 'logs' to 'history'")
```

```python
import os
import shutil

input_path = os.path.join("lab_data", "input", "notes.txt")
output_path = os.path.join("lab_data", "output", "summary.txt")

# 5. Create notes.txt with 5 lines
lines = ["Line 1\n", "Line 2\n", "Line 3\n", "Line 4\n", "Line 5\n"]
with open(input_path, "w") as f:
    f.writelines(lines)

# 6. Read notes.txt and print total number of lines
with open(input_path, "r") as f:
    content = f.readlines()
    print(f"Total number of lines: {len(content)}")

# 7. Copy notes.txt to summary.txt
# Ensure output folder exists first
os.makedirs(os.path.dirname(output_path), exist_ok=True)
shutil.copy(input_path, output_path)
print("File copied to summary.txt")
```
