```python
import numpy as np

# 1. Matrix Slicing
matrix = np.array([
    [5, 8, 2, 7],
    [1, 9, 4, 3],
    [6, 0, 11, 10],
    [12, 13, 14, 15]
])

print("The second row:", matrix[1, :])
print("The third column:", matrix[:, 2])
print("Sub-table:\n", matrix[1:3, 1:3])

# 2. Students' Scores
scores = np.array([
    [70, 80, 90],
    [60, 75, 85],
    [88, 92, 79],
    [55, 65, 72]
])

scores[:, 1] += 5
scores[0, :] = [100, 100, 100]
scores[:, 2] = 0

# 3. Operations
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
v = np.array([2, 4])

print("A * B:\n", A * B)
print("A @ B:\n", A @ B)
print("A * v:\n", A * v)

# 4. Unique Values
np.random.seed(0) 
rand_matrix = np.random.randint(-5, 6, size=(4, 4))
val, counts = np.unique(rand_matrix, return_counts=True)

a = np.array([[1, 0, 0], [1, 0, 0], [2, 3, 4]])
print("Unique rows:\n", np.unique(a, axis=0))

# 5. Advanced Functions
x = np.array([2, 0, 1, 5, 4, 1, 9])
indices = np.argsort(x)
print("Sorted x:", x[indices])

min_loc = np.unravel_index(A.argmin(), A.shape)
print("Min location:", min_loc)

print("Where x > 3:", np.where(x > 3))
print("Split x:", np.split(x, [2, 5]))

x_exp = np.expand_dims(x, axis=0)
print("Squeezed shape:", np.squeeze(x_exp).shape)
```
